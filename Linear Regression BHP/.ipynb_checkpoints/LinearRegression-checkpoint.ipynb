```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("Boston.csv")
data.head(10)
```

```python
data.describe()
```

### Removing Outliers

```python
upper = data.crim.quantile(0.95)
print(upper)

boolean = data.crim<upper
data = data[boolean]
data

upper = data.medv.quantile(0.99)
print(upper)

boolean = data.medv<upper
data = data[boolean]
data

lower = data.medv.quantile(0.01)
print(lower)

boolean = data.medv>lower
data = data[boolean]
data
```

### Simple Regression

```python
x = np.array(data.crim).reshape(-1, 1)
y = np.array(data.medv).reshape(-1, 1)
```

```python
# Splitting
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y)
```

```python
linAlg = LinearRegression()
linAlg.fit(x_train, y_train)
slope = linAlg.coef_
inter = linAlg.intercept_
print(slope)
print(inter)
```

```python
x_axis = np.arange(2).reshape(-1, 1)
y_plots = slope * x_axis + inter
plt.plot(x_axis, y_plots)
plt.scatter(x, y)
plt.show()
```
