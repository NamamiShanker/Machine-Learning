```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
columns = ['type', 'alchohol', 'malic_acid', 'ash', 'alcalinity', 'magnesium', 'phenols_total', 'flavanoids', 'phenols', 'proanthocyanins', 'intensity_color', 'hue', 'gib', 'proline']

len(columns)
```

```python
df = pd.read_csv('wine.data', names=columns)
df
```

```python
df.describe()
```

```python
mmscaler = MinMaxScaler()
mm = mmscaler.fit_transform(df)
mm
```

```python
df_mm = pd.DataFrame(mm)
df_mm
```

```python
sscaler = StandardScaler()
ss = sscaler.fit_transform(df)
ss
```

```python
df_ss = pd.DataFrame(ss)
df_ss
```
