```python
import numpy as np
import pandas as pd
```

```python
file = pd.read_csv('heights.csv')
file.head(20)
```

```python
upper_quant = file['height'].quantile(0.95) 
bool_ser = file['height']<upper_quant # Creates a series of boolean value
print(bool_ser)
file[bool_ser]
```

```python
lower_quant = file['height'].quantile(0.05)
bool_ser = file['height']>lower_quant
file[bool_ser]
```

```python
data = pd.read_csv('bhp.csv')
data.head(20)
```
